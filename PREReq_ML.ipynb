```python
#Question 3:
#Answer: The time complexity for the initial method of calculation of the median will be O(n+1).

#Method I
def find_l(arr):
    l = float('-inf')
    for n in arr:
        if n > l:
            l = n
    return l

def find_s(arr, l):
    s = float('-inf')
    for n in arr:
        if n > s and n < l:
            s = n
    return s

def sort(arr):
    ans = []
    
    while arr:
        l = find_l(arr)
        arr.remove(l)
        
        if arr:
            s = find_s(arr, l)
            arr.remove(s)
        else:
            s = None

        ans.append(l)
        if s is not None:
            ans.append(s)

    return ans

# Sorting array:
array = [5, 1, 2, 7, 4, 6, 3]
result = sort(array)
print("Sorted Array:", result)
#Finding median
median = 0
k = len(result)
if k%2 == 0:
    median = (result[int(k/2)] + result[int((k/2) + 1)])/2
else:
    median = result[int((k-1)/2)]
print(median)




#Method of sorting an array using Selection sort
arr = [5,1,2,7,4,6,3]
for i in range (7):
    for j in range (7):
        if arr[i] > arr[j]:
            temp = arr[i]
            arr[i] = arr[j]
            arr[j] = temp
#Finding median of the sorted array
l = len(arr)
median = 0
if l%2 == 0:
    median = (arr[l/2] + arr[(l+1)/2])/2
else:
    median = arr[int((l-1)/2)]
print(arr)
print(median)
        


#Method of Using numpy.median

import numpy as np

arr_np = [5,1,2,7,4,6,3]
median_np = np.median(arr_np)
print(median_np)

#Method of finding the median by numpy is fastest
```

```python
#Question 6:

import sympy
from sympy import *
x = symbols('x')
y = symbols('y')
f = x*x*y + y*y*y*sin(x)

#Derivative calculation
df_x = diff(f, x)
df_y = diff(f, y)

df_x0 = np.vectorize(lambda x_val, y_val: df_x.subs({x: x_val, y: y_val}))
df_y0 = np.vectorize(lambda x_val, y_val: df_y.subs({x: x_val, y: y_val}))

#Random values
x_val = 1
y_val = 1

print("The derivative with respect to x is", df_x) 
print("The derivative with respect to y is", df_y) 

result_x = df_x0(x_val, y_val)
result_y = df_y0(x_val, y_val)

print(result_x) 
print(result_y)
```
